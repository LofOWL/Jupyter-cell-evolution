```python
total,identical,same_position,not_same_position = list(),list(),list(),list()
```

```python
def identical_count(ouptut):    
    ce = CellEvolutionAnalyser(output)
    total.append(len(ce.lists))
    identical_v,same_position_v,not_same_position_v = 0,0,0
    for i in ce.lists:
        if ce.is_identical(i):
            identical_v +=  1
            if ce.is_same_position(i):
                same_position_v += 1
            else:
                not_same_position_v += 1
    identical.append(identical_v)
    same_position.append(same_position_v)
    not_same_position.append(not_same_position_v)
```

```python
for repo in ['18F#tech-talks']: 
    try:
        rc = RepoCheck(repo)
        _,_,outputs = rc.group()
        for output in outputs:
            if len(output) >=2:
                a_mapping_names,names = get_mappings(output)
                try:
                    mappings = [CellMapping(f'{mapping_path}/{mapping_name}') for mapping_name in a_mapping_names] 
                    output = cell_evolution_main(names,mappings)
                    identical_count(output)
                except Exception as e:
                    print(f'{repo} {e}')
    except Exception as e:
        print(f"{repo} {e}")
```
